```python
import pandas as pd
import mysql.connector
#把数据从CSV插进MySQL表，用python连接数据库
conn = mysql.connector.connect(host='localhost', user='root', password='123456', database='ecommerce_db')
cursor = conn.cursor()
orders = pd.read_csv('olist_orders_dataset.csv')
for index, row in orders.iterrows():
    cursor.execute("INSERT INTO orders (order_id, customer_id, order_status, order_purchase_timestamp) VALUES (%s, %s, %s, %s)",
    (row['order_id'], row['customer_id'], row['order_status'], row['order_purchase_timestamp']))
items = pd.read_csv('olist_order_items_dataset.csv')
for index, row in items.iterrows():
    cursor.execute("INSERT INTO items (order_id, product_id, price) VALUES (%s, %s, %s)",
    (row['order_id'], row['product_id'], row['price']))
customers = pd.read_csv('olist_customers_dataset.csv')
for index, row in customers.iterrows():
    cursor.execute("INSERT INTO customers (customer_id, customer_state) VALUES (%s, %s)",
    (row['customer_id'], row['customer_state']))
conn.commit()
conn.close()
```

```python
#用SQL语句在数据库中查询月销售和客户分布
conn = mysql.connector.connect(host='localhost', user='root', password='123456', database='ecommerce_db')
cursor = conn.cursor()
cursor.execute("SELECT MONTH(order_purchase_timestamp) AS month, COUNT(*) AS sales FROM orders GROUP BY month")
sales_by_month = cursor.fetchall()
print(sales_by_month)
cursor.execute("SELECT c.customer_state, COUNT(*) AS count FROM orders o JOIN customers c ON o.customer_id = c.customer_id GROUP BY c.customer_state")
customer_by_state = cursor.fetchall()
print(customer_by_state)
conn.close()
```
